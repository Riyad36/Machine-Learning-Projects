```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('../input/insurance-data/insurance_data.csv')
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance, test_size=0.2, random_state=30)
```

```python
len(X_train)
```

```python
len(X_test)
```

I will scale the feature age

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_train.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
X_test_scaled
```

Now I will build a simple neural network with tensorflow with a single layer

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy']
             )

model.fit(X_train_scaled, y_train, epochs=5000)
```

```python
model.evaluate(X_test_scaled, y_test)
```

Now We will build the model from stracth using python(not tensorflow)


```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
def prediction(age, affordibility):
    weigthed_sum = coef[0]*age + coef[1]*affordibility + intercept
    
    return sigmoid(weigthed_sum)
```
